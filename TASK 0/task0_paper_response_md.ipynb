1. The authors were trying to improve sequence-to-sequence tasks such as machine translation, where earlier models mainly used recurrent neural networks (RNNs) and LSTMs. These older models processed words one at a time, which made training slow and hard to parallelize. They also couldn’t capture long-range relationships between words in long sentences properly. The paper gives an alternative for these recurrent and convolutional architectures as the Transformer, a model based entirely on attention mechanisms. This allowed much faster training, better parallel computation, and improved performance on translation tasks.

2. Self-attention computes how relevant other words in a sequence are to a particular word. Imagine every word creates three vectors:
*   a query (what it is looking for),
*   a key (what information it contains),
*   and a value (the actual information it can provide).

For one word, we compare its query with the keys of all other words using dot products. The higher the similarity, the more attention that word receives. After applying softmax, these attention scores become weights, and we take a weighted sum of the value vectors.
The result is a new representation of the word that already contains relevant context from the whole sentence. So instead of processing words sequentially like an RNN, self-attention allows every word to directly gather information from all other words at once.
